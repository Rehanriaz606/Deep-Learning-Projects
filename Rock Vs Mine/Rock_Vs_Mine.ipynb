```python
import pandas as pd
```

```python
data=pd.read_csv("/content/Copy of sonar data.csv")
```

```python
data.head()
```

```python
data.R.value_counts()
```

```python
dic={'M':0,'R':1}
```

```python
data.replace({'R':dic},inplace=True)
```

```python
data.head()
```

```python
data.R.value_counts()
```

```python
x= data.drop("R",axis=1)
```

```python
y=data.R
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data,train_label,test_label=train_test_split(x,y, test_size=0.15,stratify=y)
```

```python
from tensorflow.keras import models
from tensorflow.keras.layers import  *
model= models.Sequential()
```

```python
model.add(Dense(120,activation="relu",input_shape=(train_data.shape[1],)))
model.add(Dense(12,activation="tanh"))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['acc'])
```

```python
model.fit(train_data,train_label,epochs=50)
```

```python
model.evaluate(test_data,test_label)
```
